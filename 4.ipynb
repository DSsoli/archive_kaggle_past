```python
print(train['count'].quantile(0.99))
print(train['count'].quantile(0.1))
print(train['count'].mean())
print(train['count'].median())
print(train['count'].std())
```

```python
high = train['count'].quantile(0.99)
low = train['count'].quantile(0.1)

train = train[(high > train['count']) & (train['count'] > low) ]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
train
```

```python
df_cont = train[[
    'hour_bef_temperature', 'hour_bef_windspeed', 'hour_bef_humidity', 'hour_bef_visibility',
    'hour_bef_ozone', 'hour_bef_pm10', 'hour_bef_pm2.5'
]]
```

```python
df_cont = pd.DataFrame(scaler.fit_transform(df_cont), index=df_cont.index, columns=df_cont.columns)
```

```python
df_cont
```

```python
df_cat = train[['hour', 'hour_bef_precipitation']]
df_cat = pd.get_dummies(data=df_cat, columns=df_cat.columns)
```

```python
df_cat
```

```python
x = pd.concat([df_cat, df_cont], axis=1)
y = train['count']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True, random_state=123)
```

```python
x_train = x_train.sort_index()
y_train = y_train.sort_index()
x_test = x_test.sort_index()
y_test = y_test.sort_index()
```

```python
baseModel = LinearRegression().fit(x_train, y_train)
```

```python
my.coef(baseModel.coef_, x_train)
```

```python
yhat = baseModel.predict(x_train)
print(my.adjustedR2(x_train, y_train, yhat))
yhat = baseModel.predict(x_test)
print(my.adjustedR2(x_test, y_test, yhat))
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
print(mean_absolute_error(y_test, yhat))
print(mean_squared_error(y_test, yhat))
```

```python
resid = y_test - yhat
plt.plot(resid)
plt.show()
```

```python
resid = y_test - yhat
sns.histplot(resid)
plt.show()
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV, KFold
```

```python
folds = KFold(n_splits=10, shuffle=True)
```

```python
elasticModel = ElasticNet()
params = {
    'l1_ratio': [0.0, 0.3, 0.5, 0.8, 1.0],
    'alpha': [0.001, 0.01, 0.1, 1, 10, 100, 100]
}
search = GridSearchCV(elasticModel, param_grid=params, cv=folds, scoring='r2')
search.fit(x_train, y_train)
```

```python
search.best_score_
```

```python
search.best_params_
```

```python
model = search.best_estimator_
```

```python
yhat = model.predict(x_test)
my.adjustedR2(x_test, y_test, yhat)
```

```python
yhat = search.predict(x_test)
my.adjustedR2(x_test, y_test, yhat)
```

```python
my.coef(search.best_estimator_.coef_, x_train)
```

```python
resid = y_test - yhat
sns.histplot(resid)
plt.show()
```
